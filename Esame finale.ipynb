2.a
Da uno sguardo veloce alle colonne che ci serviranno per le analisi richieste, si nota subito che ci sono dei valori mancanti nella colonna continent. Facciamo un riscontro numerico dei valori NaN e decidiamo come procedere

```python
covid["continent"].isna().sum()
```

Vediamo se riusciamo a recuperare qualche dato.
Conoscendo le location possiamo ricavare il continente di appartenenza, quindi verifichiamo i valori di location corrispondenti ai Nan di continent

```python
continent_null=covid[covid["continent"].isna()][["continent","location"]]
continent_null
```

Interessante! In location abbiamo valori di continent. Controlliamo quali altri valori ha la variabile location nel db filtrato per valori Nan di continent

```python
continent_null["location"].unique()
```

```python
#sostiuiamo i valori mancanti di continent con i rispettivi valori di location
continents= ["Africa", "Asia", "Europe","North America", "Oceania", "South America"]

for cont in continents:
    filtro_location= covid["location"]== cont
    filtro_nan= covid["continent"].isna()
    filtro_rep= filtro_location & filtro_nan
    covid.loc[filtro_rep,"continent"]= cont
```

```python
#verifichiamo se le sostituzioni sono state utili
covid["continent"].isna().sum()
```

In mumero di valori nulli è diminuito, ma restano 16.454 NaN che potrebbero essere di intralcio per la nostra analisi! Quindi mi credo un nuovo db senza i NaN.
Dalla cella [6] però avevo notato che il type della colonna date è object, quindi prima trasformo il type in datatime, e poi mi creo il nuovo db

```python
covid["date"]= pd.to_datetime(covid["date"])
```
